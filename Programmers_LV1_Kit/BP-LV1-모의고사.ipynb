```python
#1  : 12345
#2   21 23 24 25
#3  : 33 11 22 44 55

def solution(answers):
    score = [0, 0, 0]
    answer = []
    c1 = [1, 2, 3, 4, 5]
    c2 = [2, 1, 2, 3, 2, 4, 2, 5]
    c3 = [3, 3, 1, 1, 2, 2, 4, 4, 5, 5]
    for i in range(len(answers)):
        if answers[i] == c1[i%5]:
            score[0] += 1
        if answers[i] == c2[i%8]:
            score[1] += 1
        if answers[i] == c3[i%10]:
            score[2] += 1
    print(score)
    for i in range(len(score)):
        if score[i] == max(score):
            answer.append(i+1)
    return sorted(answer)
```
