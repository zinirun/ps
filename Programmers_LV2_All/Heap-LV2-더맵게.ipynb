```python
def solution(scoville, K):
    cnt=0
    while scoville:
        mix=scoville.pop(scoville.index(min(scoville)))+scoville.pop(scoville.index(min(scoville)))*2
        scoville.insert(0,mix)
        cnt+=1
        if mix>=K: break
    return cnt if scoville[0]>=K else -1
```

```python
print(solution([1,2,3,9,10,12],7))
```

```python
import heapq as h
def solution(scoville, K):
    s = []
    cnt=0
    for i in scoville: h.heappush(s, i)
    while len(s)>1:
        if s[0]>=K: break
        h.heappush(s, h.heappop(s)+h.heappop(s)*2)
        cnt+=1
    return cnt if s[0]>=K else -1
```

```python
print(solution([1,2,3],30))
```
