```python
def solution(s):
    p = []
    for i in s:
        if i == '(':
            p.append('(')
        else:
            if len(p) > 0: p.pop()
            else: return False
    return len(p) == 0
```

```python
def check(p):
    stack = []
    try:
        for i in p:
            if i == '(':
                stack.append('(')
            else:
                stack.pop()
        return True
    except:
        return False

def divide(p):
    count = [0, 0]
    for i in p:
        if i == '(':
            count[0] += 1
        else:
            count[1] += 1
        if count[0] == count[1]:
            break
    return p[:sum(count)], p[sum(count):]

def convert(u):
    temp = ''
    for i in u:
        if i == '(':
            temp += ')'
        else:
            temp += '('
    return temp

def solution(p):
    answer = ''

    while p:
        u, p = divide(p)
        if check(u):
            answer += u
        else:
            answer += '(' + solution(p) + ')' + convert(u[1:-1])
            break

    return answer
```
