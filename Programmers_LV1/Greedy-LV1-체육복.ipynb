```python
# n:   
# lost:  
# reserve:  
# lost, reserve   
def solution(n, lost, reserve):
    survive = []
    same = 0
    lost.sort()
    reserve.sort()
    for i in lost:
        for j in reserve:
            if i == j:
                del reserve[reserve.index(j)]
                survive.append(i)
                same += 1
                break
            if abs(i-j) == 1:
                del reserve[reserve.index(j)]
                survive.append(i)
                break
    print(survive)
    return n - len(lost) + len(survive) - same
    
solution(5, [2, 3], [3, 4])
```

```python
def solution(n, lost, reserve):
    _reserve = [r for r in reserve if r not in lost]      #     list
    _lost = [l for l in lost if l not in reserve]         #   1    0 

    for r in _reserve:                                    #    -1 +1     
        f = r - 1                                         
        b = r + 1
        if f in _lost:                                    #  lost list 
            _lost.remove(f)
        elif b in _lost:
            _lost.remove(b)
    return n - len(_lost)
```
