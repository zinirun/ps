```python
def solution(phone_book):
    phone_book.sort()
    phone_book = list(map(str,phone_book))
    for pattern in phone_book:
        for i, j in enumerate(phone_book[:-1]):
            if pattern != j and j.find(pattern) == 0:
                return False
    return True

solution([123,456,789])
```
