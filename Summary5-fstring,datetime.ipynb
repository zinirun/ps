```python
# 3.6 Version
age = 10
name = 'zini'
print('제 나이는', age, '입니다')
print('제 나이는 {} 입니다, 제 이름은 {}입니다'.format(age, name))
print(f'제 나이는 {age} 입니다. 제 이름은 {name}입니다.') # fstring 용법
```

```python
for i in range(2, 10):
    for j in range(1, 10):
        print(f'{i} X {j} = {i*j}')
```

```python
print(f'{"hello":<10}') #10글자 기준 왼쪽 정렬
print(f'{"hello":>10}') #오른쪽 정렬
print(f'{"hello":^10}') #중간 정렬
```

```python
#공백 채우기
print(f'{"hello":!<10}') #10글자 기준 왼쪽 정렬
print(f'{"hello":=>10}') #오른쪽 정렬
print(f'{"hello":~^10}') #중간 정렬
```
